```python
cat1_list = []

for list_data in new_mod_list:
    new_list_data = []
    temp_list = []
    nan_appended = False  # Flag to track whether np.nan has been appended
    flag = 0
    for string in list_data:
        if pd.notna(string):
            if 'Fiction' in string and 'Fiction' not in temp_list:
                temp_list.append('Fiction')
                if(len(string)>7):
                  new_list_data.append(string)
                flag=1
            elif 'Nonfiction' in string and 'Nonfiction' not in temp_list:
                temp_list.append('Nonfiction')
                if(len(string)>10):
                  new_list_data.append(string)
                flag=1
            else:
                new_list_data.append(string)
                # print('came here and new list data is:', new_list_data)
        else:
            new_list_data.append(string)
            if not nan_appended:  # Append np.nan only once
                cat1_list.append(np.nan)
                nan_appended = True
    if not nan_appended:
      cat1_list.append(temp_list)

    # Update list_data with the new list
    list_data[:] = new_list_data

# # Display the resulting lists
# print("cat1_list:", cat1_list)
# print("new_mod_list:", new_mod_list)
```
