```python
# Dataset preprocessing
from nltk.stem.porter import PorterStemmer # for stemming
ps = PorterStemmer()
corpus = []
try_encodings = ['utf-8', 'latin-1', 'iso-8859-1', 'cp1252']
# try_encodings = ['utf-8']
# Download stopwords for both English and Spanish
stopwords_english = set(stopwords.words('english'))
# stopwords_spanish = set(stopwords.words('spanish'))

# Combine the stopwords lists
# stopwords_combined = stopwords_english.union(stopwords_spanish)

for i in range(len(common_isbn_df)): # len(common_isbn_df) -> length of merged doc
  file_path = os.path.join(description_data, common_isbn_df['Description'][i])
  # print(i)
  # Open the file in read and write mode
  for encoding in try_encodings:
    try:
        with open(file_path, 'r+', encoding=encoding) as file:
            file_contents = file.read() # reading contents of file
        break
    except UnicodeDecodeError:
        continue
  review = re.sub('[^a-zA-Z]', ' ', file_contents) # means, replacing all the characters, apart from a-z and A-Z, with '' blank space
  review = review.lower()
  review = review.split() # splitting - default delimiter white space
  # now review will be having the list of words insisde that specific sentence

  stemmed_words = []
  for word in review:
    if word not in stopwords_english:
      word = ps.stem(word)
      stemmed_words.append(word)
  review = stemmed_words
  review = ' '.join(review)
  corpus.append(review)
```

```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder

# Create the CountVectorizer
vectorizer = CountVectorizer()

# Fit and transform the corpus into a bag-of-words matrix
X = vectorizer.fit_transform(corpus).toarray()
feature_names = vectorizer.get_feature_names_out()

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.33, random_state=42)

# Encode class labels using LabelEncoder
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# Check the mapping of original class names to encoded labels
class_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
print("Class Mapping:", class_mapping)

# Calculate class weights to address class imbalance
class_weights = len(y_train_encoded) / (2 * np.bincount(y_train_encoded))

# XGBoost classifier with class weights
# the calculated class weights to the scale_pos_weight parameter of the XGBoost classifier.
# This parameter adjusts the balance between positive and negative classes during training.
model = XGBClassifier(scale_pos_weight=class_weights[1])

# Train the model on the training data
model.fit(X_train, y_train_encoded)

# Make predictions on the test data
predictions = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test_encoded, predictions)
conf_matrix = confusion_matrix(y_test_encoded, predictions)
classification_rep = classification_report(y_test_encoded, predictions)

# Display the results
print("Accuracy:", accuracy)
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(classification_rep)
```
